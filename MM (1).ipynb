```python
import pandas as pd

file_path = 'C:/Users/user/OneDrive/سطح المكتب/PhiUSIIL_Phishing_URL_Dataset3.csv'  
# dataset load
df = pd.read_csv(file_path)

print(" Dataset loaded successfully!")
print(df.head())
```

```python
print("\n Columns in the dataset:")
print(df.columns.tolist())
```

```python
print("\n Descriptive Statistics (for numeric columns):")
print(df.describe())
```

```python
print("\n Unique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
print("\n Missing Values per Column:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical features into numeric
encoder = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = encoder.fit_transform(df[col])

# Split data into features (X) and target (y)
X = df.drop('label', axis=1)
y = df['label']
```

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import ElasticNet, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    matthews_corrcoef, roc_auc_score, confusion_matrix, classification_report, roc_curve
)

# Start timer
start_time = time.time()

# Load and split data
X = df.drop('label', axis=1)
y = df['label']

# ElasticNet Feature selection
elastic = ElasticNet(alpha=0.01, l1_ratio=0.5, max_iter=1000, random_state=42)
elastic.fit(X, y)
mask = np.abs(elastic.coef_) > 1e-5
X_selected = X.loc[:, mask]
selected_features = X.columns[mask]

print(f"Selected {len(selected_features)} features:")
print(selected_features.tolist())

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Base models
dt = DecisionTreeClassifier(random_state=42)
lr = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000, random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42)
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# All models + ensemble
models = {
    'Decision Tree': dt,
    'Logistic Regression (L1)': lr,
    'MLP': mlp,
    'XGBoost': xgb,
    'Ensemble (Voting)': VotingClassifier(
        estimators=[
            ('dt', dt),
            ('lr', lr),
            ('mlp', mlp),
            ('xgb', xgb)
        ],
        voting='soft'
    )
}

# Prepare to store ROC data
roc_data = {}

# Train and evaluate
for name, model in models.items():
    print(f"\n--- {name} ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        y_prob = np.zeros_like(y_test)

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(name)
    plt.show()

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    print(f"Accuracy: {acc:.4f}, F1: {f1:.4f}, Precision: {prec:.4f}, MCC: {mcc:.4f}, AUC: {auc:.4f}")

    # Print classification report
    report_dict = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report_dict).transpose()

    # Remove accuracy row (optional)
    if 'accuracy' in report_df.index:
        report_df = report_df.drop(index='accuracy')

    report_df = report_df[['precision', 'recall', 'f1-score', 'support']]
    for col in ['precision', 'recall', 'f1-score']:
        report_df[col] = report_df[col].round(4)
    report_df['support'] = report_df['support'].astype(int)

    print("\nDetailed Classification Report:")
    print(report_df)

    # Save ROC data
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_data[name] = (fpr, tpr)

# ROC Curve for all models
plt.figure(figsize=(12, 9))
for name, (fpr, tpr) in roc_data.items():
    plt.plot(fpr, tpr, linewidth=2.5, label=name)
    plt.fill_between(fpr, 0, tpr, alpha=0.2)

plt.plot([0, 1], [0, 1], 'k--', linewidth=2)
plt.title('ROC Curves for All Models', fontsize=18)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()

# End timer
print(f"\nExecution time: {time.time() - start_time:.2f} seconds")
```

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    matthews_corrcoef, roc_auc_score, confusion_matrix, classification_report, roc_curve
)

# Start timer
start_time = time.time()

# Split columns
X = df.drop('label', axis=1)
y = df['label']

# Apply L2 Feature Selection 
lr_selector = LogisticRegression(penalty='l2', solver='lbfgs', max_iter=1000, random_state=42)
selector = SelectFromModel(lr_selector, threshold='mean')
selector.fit(X, y)
X_selected = selector.transform(X)
selected_features = X.columns[selector.get_support()]

print(f"Selected {len(selected_features)} features (by L2 Regularization):")
print(selected_features.tolist())

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Define base models
dt = DecisionTreeClassifier(random_state=42)  # No L2 for DecisionTree
lr = LogisticRegression(penalty='l2', solver='lbfgs', max_iter=1000, random_state=42)  
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42)  # Removed alpha
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)  # Removed reg_lambda

# Build models
models = {
    'Decision Tree': dt,
    'Logistic Regression (L2)': lr,
    'MLP': mlp,
    'XGBoost': xgb,
    'Ensemble (Voting)': VotingClassifier(
        estimators=[
            ('dt', dt),
            ('lr', lr),
            ('mlp', mlp),
            ('xgb', xgb)
        ],
        voting='soft'
    )
}

# Prepare to store ROC data
roc_data = {}

# Train and evaluate
for name, model in models.items():
    print(f"\n--- {name} ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        y_prob = np.zeros_like(y_test)

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(name)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    print(f"Accuracy: {acc:.4f}, F1: {f1:.4f}, Precision: {prec:.4f}, MCC: {mcc:.4f}, AUC: {auc:.4f}")

    # Print detailed classification report
    report_dict = classification_report(y_test, y_pred, output_dict=True)
    report_df = pd.DataFrame(report_dict).transpose()

    if 'accuracy' in report_df.index:
        report_df = report_df.drop(index='accuracy')

    report_df = report_df[['precision', 'recall', 'f1-score', 'support']]
    for col in ['precision', 'recall', 'f1-score']:
        report_df[col] = report_df[col].round(4)
    report_df['support'] = report_df['support'].astype(int)

    print("\nDetailed Classification Report:")
    print(report_df)

    # Save ROC curve data
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_data[name] = (fpr, tpr)

# Draw ROC Curve for all models
plt.figure(figsize=(12, 9))
for name, (fpr, tpr) in roc_data.items():
    plt.plot(fpr, tpr, linewidth=2.5, label=name)
    plt.fill_between(fpr, 0, tpr, alpha=0.2)

plt.plot([0, 1], [0, 1], 'k--', linewidth=2)
plt.title('ROC Curves for All Models', fontsize=18)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()

# End timer
print(f"\nExecution time: {time.time() - start_time:.2f} seconds")
```
